```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import glob

```

```python
infiles = sorted(glob.glob('pdat/*'))
data = np.loadtxt(infiles[-1], skiprows = 1)
time = data[:,0]
ph = data[:,1]
cond = data[:,2]
places = np.where(np.roll(ph,1) > ph)[0]
print(infiles[-1])
```

```python
pr = np.empty((len(places),2))
for i in range(len(places)):
    if i>=1:
        probeind = (places[i-1],places[i])
        pr[i,0] = probeind[0]
        pr[i,1] = probeind[1]
dupes = np.where(pr[:,1] - pr[:,0] == 1 )[0]
pr = np.delete(pr, dupes, axis = 0)
if len(pr) > 50:
    pr = pr[-12:]
pr = pr[2:]
print(pr)
```

```python
#control conductivity
for i in range(len(pr)):
    xax = int(pr[i,0]) 
    yax = int(pr[i,1]) 
    tx = time[xax:yax:1]
    cplot = cond[xax:yax:1]
    cplot = cplot[1:]
    pplot = ph[xax:yax:1]
    pplot = pplot[1:]
    plt.plot(cplot,-pplot)
    #plt.title('Conductivity at Time ' + str(tx[0]) + 's to ' + str(tx[-1]) + 's' )
    plt.title('Typical Time Evolution')
    plt.ylabel('Probe Height (mm)')
    plt.xlabel('Conductivity (V)')
plt.savefig('Time_evolution.png')
plt.show()
#compare earlier days with later days
```

```python
infile1 = sorted(glob.glob('p2dat/*'))
data1 = np.loadtxt(infile1[0], skiprows = 1)
ti = data1[:,0]
ph1 = data1[:,1]
mix = data1[:,2]
cond1 = data1[:,3]

index = np.where(np.roll(ph1,1)!= ph1)[0]
end = len(ph1) - 1
index = np.append(index,end)
places1 = len(index)
pr1 = np.empty((places1,2))
for i in range(places1):
    if i >= 1:
        probeind1 = (index[i-1], index[i])
        pr1[i,0] = probeind1[0]
        pr1[i,1] = probeind1[1]
pr1 = pr1[1:]

#print(places1)
dupes1 = np.where(pr1[:,1] - pr1[:,0] < 100 )[0]
pr1 = np.delete(pr1, dupes1, axis = 0)
print(infile1[0])
print(len(pr1))
```

```python
#velocimiter conductivity
for i in range(len(pr1)):
    xax1 = int(pr1[i,0])
    yax1 = int(pr1[i,1])
    time = ti[xax1:yax1:1]
    con = cond1[xax1:yax1:1]
    time = time[1:]
    con = con[1:]
    plt.plot(time, con)
    plt.title('Probe 2 conductivity at Height ' + str(ph1[xax1]) + 'mm') 
    plt.xlabel('Time (s)')
    plt.ylabel('Conductivity (V)')
    plt.show()
```

```python
dirlist = sorted(glob.glob('velocitydata/03-18data/*'))
sortedfiles = np.reshape(dirlist, (-1, 2))
dim = len(sortedfiles[:,0])
gatedep = np.arange(139,(274+15), 15)
dim
```

```python
#Velocity Profiles
#gate = np.arange(0.75, 188.25, 0.75)
gate = np.arange(7.5, 75.75, 0.75)
averages = np.empty((dim))
gh = np.empty((dim))
for i in range(dim):
    veldata1 = pd.read_csv(sortedfiles[i,0], sep = '\t', skiprows = 5, engine = 'python')
    veldata2 = pd.read_csv(sortedfiles[i,1], sep = '\t', skiprows = 5, engine = 'python')
    veldata1.drop(veldata1.columns[[-1,-2,-3,-4,-5]], axis=1, inplace=True)
    veldata2.drop(veldata2.columns[[-1,-2,-3,-4,-5]], axis=1, inplace=True)
    dim2 = np.arange(0,10,1)
    veldata1.drop(veldata1.columns[dim2], axis=1, inplace=True)
    veldata2.drop(veldata2.columns[dim2], axis=1, inplace=True)
    veldata1.loc['avg'] = veldata1.mean()
    veldata2.loc['avg'] = veldata2.mean()
    avvel1 = np.array(veldata1.loc['avg'])
    avvel2 = np.array(veldata2.loc['avg'])
    veldata1.loc['std'] = veldata1.std()
    veldata2.loc['std'] = veldata2.std()
    velstd1 = np.array(veldata1.loc['std'])
    velstd2 = np.array(veldata2.loc['std'])
    vel = np.append(veldata1.loc['avg'],veldata2.loc['avg'])
    mv = np.mean(vel)
    st = np.std(vel)
    avvel = (avvel1+avvel2) * 0.5
    avvel = avvel[:-1]
    avvel0 = avvel[:len(gate)]
    m1 = np.mean(avvel1)
    m2 = np.mean(avvel2)
    sd1 = np.std(avvel1)
    sd2 = np.std(avvel2)
    relerr = 0.5 * ((sd1/m1) + (sd2/m2))
    sterr = (m1+m2)* 0.5 * relerr
    #otherthing = (avvel1+avvel2)*0.5*relerr
    #otherthing = otherthing[:-1]
    averages[i] = np.mean(avvel)
    gdep = gatedep[i]
    gh[i] = gdep
    print(mv,st)
    plt.plot(avvel0, gate)
    plt.title('Velocity Profile at ' + str(gdep) + 'mm')
    plt.xlabel('Velocity (mm/s)')
    plt.ylabel('Gate Depth (mm)')
    plt.show()
    #print(len(velstd1))
   # print(veldata1)
#Mixing Profile
g = gh * -1
plt.scatter(averages, g)
plt.xlabel('Average Velocity (mm/s)')
plt.ylabel('Height (mm)')
plt.title('Upchannel velocity Profile')
plt.show()
#trim velocity profile 10<-->60
#print(len(gate))
```
